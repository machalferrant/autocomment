```python
import re
import pandas as pd
import numpy as np

list1 = []
digit = []
spam_words = []


# «Спамовые» слова

words = ["прошу", "помогите", "помочь", "помощь", "карта", "карту", "номер", "перевести", "поторопись", 
         "подпишись", "получи", "получишь", "акция", "загляни", "посмотри", "приглашаем", "приглашаю", "подарок", 
         "хочешь", "скиньте", "скинь", "скинуть", "деньги", "бабки", "заработать", "заработай", "зарабатывай", 
         "зарабатываю", "работать", "работаю", "молю", "нуждается", "нуждаемся", "помощи", "поддержке",
         "поддержите", "зайка", "зайке", "заика", "заике", "сбор", "жизнь", "сбору", "шанс", "маски", "масок", 
         "тыс", "руб", "р", "₽", "сторис", "страница", "странице", "страничка", "страничке", "акк", "аккаунт", 
         "аккаунта", "аккаунте", "маска", "продам", "заболевание", "болезнь", "карточка", "карточку", "карточки", 
         "зайди", "зайдите", "заходи", "заходите", "зайти", "подписки", "подписка", "подписку", "лайк", "лайки", 
         "сторис", "подписывайтесь", "подпишись", "подписывайся", "паблик", "подписаться", "рублей", "подарим", 
         "подарить", "реклама", "рекламу", "уволилась", "глянь", "гляньте", "глянуть", "ушла", "ушел", "уволилась", 
         "уволился"]


# Загружаем документ – результат веб-скрейпинга

table = pd.read_csv('Truelentach.csv')
table


# Удаляем все пропущенные значения

table.dropna(inplace=True)
table

table_list = []


# Приводим данные в формат, который будет читать программа

for i in table['comments']:
    table_list.append(i)
    
table_list

table_list_2 = [[i] for i in table_list]
table_list_2

let = []

for i in table_list_2:
    for k in i:
        let.append(list(k))
print(let)


# Функция cr1: считает по буквам

def cr1(phrase):
    if len(phrase) > 3:
        a = 'ok'
    else:
        a = 'spam'
    return a

table['cr1'] = table['comments'].apply(cr1)
table


# Функция cr2: количество цифр

def cr2(phrase):
    num = []
    ser = list(phrase)
    for i in ser:
        if i.isdigit():
            num.append(i)
    if len(num) >= 5:
        a = 'spam'
    else:
        a = 'ok'
    return a

table['cr2'] = table['comments'].apply(cr2)
table


# Функция cr3: латиница

def cr3(phrase):
    if len([i for i in filter(re.compile('[A-Za-z]').match, phrase)]) > 5:
        a = 'spam'
    else:
        a = 'ok'
    return a

table['cr3'] = table['comments'].apply(cr3)
table


# Функция cr4: проверяем на символы

def cr4(phrase):
    if phrase.count('.') >= 4 or phrase.count ('@') > 0 or phrase.count('#') > 0 or phrase.count('w') >= 3 or phrase.count('!') > 1 or phrase.count('/') >= 2 or phrase.count('№') >= 1 or phrase.count('.
    ru') >= 1 or phrase.count('.com') >= 1:
        a = 'spam'
    else:
        a = 'ok'
    return a

table['cr4'] = table['comments'].apply(cr4)
table


# Функция cr5: проверяем на спамовые слова

def cr5(phrase):
    spam_words = []
    low_phrase = phrase.lower()
    comma_phrase = low_phrase.replace(',', '').replace('.', '').replace('!', '').replace('?', '')
    find_spam_words = comma_phrase.split()

    for i in find_spam_words:
        if i == i in words:
            spam_words.append(i)

    if len(spam_words) == 0:
        a = 'ok'
    elif 0 <= len(spam_words) <= 1:
        a = 'spam'
    else:
        a = 'very spam'
    return a

table['cr5'] = table['comments'].apply(cr5)
table


# Считаем very spam

table['spam'] = table['cr5'].apply(lambda x: '100' if x == 'very spam' else 0)
table


# Считаем остальные коэффициенты

table['auto'] = list(table['cr1'.format(i)].apply(lambda x: round(x.count('spam') * 100 / 5)))
for i in range(2, 5+1):
    table['auto'] += list(table['cr{}'.format(i)].apply(lambda x: round(x.count('spam') * 100 / 5)))
table

table.to_excel('truelentach_scrape.xlsx')
```
